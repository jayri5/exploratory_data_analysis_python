# As is clear from above, this dataset contains the cast details inside the "cast" column alongwith many other details that are not very useful for analysis. Also, the cast column is an object-type column that contains values in the form of dictionary but inside list, hence extracting only the artists' name is bit challenging. This work is done below for 2 movies (same code can be automated for al the movies)

```python
df[df['title'] == 'Titanic']
```

```python
import re
test_str = df['cast'][25]

start = df['cast'][25].find('name')    
end = df['cast'][25].find('order') 
target_column = df['cast'][1][start:end]

test_sub = target_column


res = [i.start() for i in re.finditer("name", test_str)]
final = [0]*len(res)
for i in range(len(res)):
    final[i] = df['cast'][25][res[i]:res[i]+25]
final    
```
